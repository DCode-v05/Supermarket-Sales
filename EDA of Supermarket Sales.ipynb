```python
#Importing Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Importing The File

mar=pd.read_csv("../EDA/supermarket_sales.csv")
mar.head()
```

```python
#Displaying The Column Details

mar.info()
```

```python
#Checking Any Null Values

mar.isnull()
```

```python
#Droping null Values

mar=mar.dropna()
mar.head()
```

```python
#Modifying The Data in The File

mar["Total"]=mar["Total"].round()
mar["gross margin percentage"]=mar["gross margin percentage"].round(2)
mar["gross income"]=mar["gross income"].round(2)
mar["Gender"]=mar["Gender"].str.replace("Female","F")
mar["Gender"]=mar["Gender"].str.replace("Male","M")
mar.head()
```

```python
#Describing The Data

mar.describe(include="all").T
```

```python
#Seperating String Values from File

marstr=mar.select_dtypes(include=["object"])
marstr.head()
```

```python
#Seperating Number Values from File

marnum=mar.select_dtypes(include=["int64","float64"])
marnum.head()
```

```python
#Plotting the Analysis About Cogs of Products

fig , ax = plt.subplots(3,2,figsize=(30,20))

fig.suptitle("Analysis About Cogs of Products" , fontsize=30 , fontweight="bold");

ax[0][0].bar(mar["Branch"],mar["cogs"] , color="g");
ax[0][0].set(xlabel="Branch" , ylabel="Cogs" , title="Branch Vs Cogs");

ax[0][1].bar(mar["Quantity"],mar["cogs"], color="r");
ax[0][1].set(xlabel="Quantity" , ylabel="Cogs" , title="Quantity Vs Cogs");

ax[1][0].bar(mar["Product line"],mar["cogs"], color="b");
ax[1][0].set(xlabel="Product Line" , ylabel="Cogs" , title="Product Line Vs Cogs");

ax[1][1].bar(mar["City"],mar["cogs"], color="y");
ax[1][1].set(xlabel="City" , ylabel="Cogs" , title="City Vs Cogs");

ax[2][0].bar(mar["Payment"],mar["cogs"], color="c");
ax[2][0].set(xlabel="Payment" , ylabel="Cogs" , title="Payment Vs Cogs");

ax[2][1].bar(mar["Customer type"],mar["cogs"], color="k");
ax[2][1].set(xlabel="Customer Type" , ylabel="Cogs" , title="Customer Type Vs Cogs");

fig.savefig("EDA1.png")
```

```python
#Plotting the Analysis About Gross Income of Products

fig , ax = plt.subplots(3,2,figsize=(30,20))

fig.suptitle("Analysis About Gross Income of Products" , fontsize=30 , fontweight="bold");

ax[0][0].bar(mar["Branch"],mar["gross income"], color="g");
ax[0][0].set(xlabel="Branch" , ylabel="Gross Income" , title="Branch Vs Gross Income");

ax[0][1].bar(mar["Quantity"],mar["gross income"], color="r");
ax[0][1].set(xlabel="Quantity" , ylabel="Gross Income" , title="Quantity Vs Gross Income");

ax[1][0].bar(mar["Product line"],mar["gross income"], color="b");
ax[1][0].set(xlabel="Product Line" , ylabel="Gross Income" , title="Product Line Vs Gross Income");

ax[1][1].bar(mar["City"],mar["gross income"], color="y");
ax[1][1].set(xlabel="City" , ylabel="Gross Income" , title="City Vs Gross Income");

ax[2][0].bar(mar["Payment"],mar["gross income"], color="c");
ax[2][0].set(xlabel="Payment" , ylabel="Gross Income" , title="Payment Vs Gross Income");

ax[2][1].bar(mar["Customer type"],mar["gross income"], color="k");
ax[2][1].set(xlabel="Customer Type" , ylabel="Gross Income" , title="Customer Type Vs Gross Income");

fig.savefig("EDA2.png")
```
